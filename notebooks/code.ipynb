```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
img1 = cv2.imread("tartaruga1A.jpg")
img_gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img_gray1=cv2.GaussianBlur(img_gray1, (13, 13), 0)
img_gray1=cv2.adaptiveThreshold(img_gray1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 4)

img2 = cv2.imread("tartaruga1B.jpg")
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img_gray2=cv2.GaussianBlur(img_gray2, (13, 13), 0)
img_gray2=cv2.adaptiveThreshold(img_gray2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 4)
```

```python
edges1 = cv2.Canny(img_gray1, 30, 90, 3, L2gradient = False)
edges2 = cv2.Canny(img_gray2, 30, 90, 3, L2gradient = False)
```

```python
contours1, hierarchy1 = cv2.findContours(edges1, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

contours2, hierarchy2 = cv2.findContours(edges2, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

```python
cv2.drawContours(img1, contours1, -1, (0, 255, 0), 2)
plt.imshow(img1)
```

```python
cv2.drawContours(img2, contours2, -1, (0, 255, 0), 2)
plt.imshow(img2)
```
